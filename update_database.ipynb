```python
import mysql.connector
import random
import time

"""
這個程式會定期更新咖啡廳的人潮數據
並將更新後的數據打印到控制台。
首先用 mysql -u root -p -h localhost -P 3306 進到sql系統(import_cafe_to_db怎麼設定就輸甚麼)
USE <DB_final>; 檢查資料庫是否存在
SET GLOBAL event_scheduler = ON; 重要!!!開啟事件排程器
更改config執行程式碼
"""


# 資料庫連線設定
config = {
    "host": "localhost",
    "user": "root",
    # g 輸入密碼
    "password": "your_password_here",
    "database": "DB_final",
    "port": 3306
}


def update_crowd_and_print():
    try:
        conn = mysql.connector.connect(**config)
        cursor = conn.cursor()

        cursor.execute("SELECT id, current_crowd FROM Cafe")
        cafes = cursor.fetchall()

        for cafe_id, old_crowd in cafes:
            delta = random.randint(-10, 10)
            new_crowd = max(0, old_crowd + delta)
            cursor.execute(
                "UPDATE Cafe SET current_crowd = %s WHERE id = %s",
                (new_crowd, cafe_id)
            )

        conn.commit()

        cursor.execute("SELECT id, name, current_crowd FROM Cafe")
        print("📊 更新後的人潮資料：")
        for row in cursor.fetchall():
            print(f"  ☕️ {row[1]} (ID: {row[0]}) -> 人潮：{row[2]}")

    except mysql.connector.Error as err:
        print("❌ SQL 錯誤：", err)
    finally:
        if conn.is_connected():
            cursor.close()
            conn.close()


if __name__ == "__main__":
    while True:
        print("🔄 執行人潮更新...")
        update_crowd_and_print()
        print("⏳ 等待 5 分鐘...\n")
        time.sleep(5 * 60)
```
